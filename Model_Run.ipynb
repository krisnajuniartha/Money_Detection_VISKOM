```python
from ultralytics import YOLO
import cv2
import time

# Peta class_id ke nominal uang
class_to_nominal = {
    0: "1000",
    1: "2000",
    2: "5000",
    3: "10000",
    4: "20000",
    5: "50000",
    6: "100000"
}

model = YOLO("weights/best.pt")
cam = cv2.VideoCapture(0)

if not cam.isOpened():
    raise Exception("No Camera")

# Set untuk melacak objek yang telah terdeteksi
detected_objects = set()

# Open file to write detection results
with open("detection_results.txt", "w") as file:
    while True:
        ret, image = cam.read()
        if not ret:
            break
        
        time_mulai = time.time()
        results = model.predict(image, show=True)
        waktu_proses = time.time() - time_mulai
        
        # Process each detection in results
        for detection in results[0].boxes:
            # Get class information and confidence
            class_id = int(detection.cls[0])
            confidence = detection.conf[0].item()
            
            # Check if this object has been detected before
            if class_id not in detected_objects:
                detected_objects.add(class_id)
                
                # Write nominal uang to file
                nominal_uang = class_to_nominal.get(class_id, "Unknown")
                file.write(f"{nominal_uang}\n")
                file.flush()  # Ensure the data is written to the file
        
        print(f"Waktu: {waktu_proses}")
        
        _key = cv2.waitKey(1)
        if _key == ord('q'):
            break

cam.release()
cv2.destroyAllWindows()
```

```python
from ultralytics import YOLO
import cv2
import time
import yaml

# Baca file YAML untuk mendapatkan daftar nama kelas
with open('data.yaml', 'r') as file:
    data = yaml.safe_load(file)

class_names = data['names']

# Buat peta class_id ke nominal uang berdasarkan class_names
class_to_nominal = {idx: class_names[idx] for idx in range(len(class_names))}

model = YOLO("weights/best.pt")
cam = cv2.VideoCapture(0)

if not cam.isOpened():
    raise Exception("No Camera")

# Set untuk melacak objek yang telah terdeteksi
detected_objects = set()

# Open file to write detection results
with open("results_matauang.txt", "w") as file:
    while True:
        ret, image = cam.read()
        if not ret:
            break
        
        time_mulai = time.time()
        results = model.predict(image, show=True)
        waktu_proses = time.time() - time_mulai
        
        # Process each detection in results
        for detection in results[0].boxes:
            # Get class information and confidence
            class_id = int(detection.cls[0])
            confidence = detection.conf[0].item()
            
            # Check if this object has been detected before
            if class_id not in detected_objects:
                detected_objects.add(class_id)
                
                # Write nominal uang to file
                nominal_uang = class_to_nominal.get(class_id, "Unknown")
                file.write(f"{nominal_uang}\n")
                file.flush()  # Ensure the data is written to the file
        
        print(f"Waktu: {waktu_proses}")
        
        _key = cv2.waitKey(1)
        if _key == ord('q'):
            break

cam.release()
cv2.destroyAllWindows()
```
